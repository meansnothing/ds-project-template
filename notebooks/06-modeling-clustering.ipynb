```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
import numpy as np

%cd ..
```

```python
df = pd.read_csv('data/processed/genshin_character_lv_90.csv')
df = df[~df['Base ATK'].isna()]
```

```python
kmeans = KMeans(n_clusters=4)
 
# Predict the labels of clusters.
label = kmeans.fit_predict(df[['Base ATK','Base DEF','Base HP']])
df['label'] = label

# Getting unique labels
u_labels = np.unique(label)
 
# Plotting the results
fig = plt.figure(figsize = (10, 7))
ax = plt.axes(projection ="3d")
for i in u_labels:
    ax.scatter3D(df.loc[df.label==i, 'Base ATK'] , df.loc[df.label==i, 'Base DEF'], df.loc[df.label==i, 'Base HP'], label = i)
plt.legend()
plt.show()
```
